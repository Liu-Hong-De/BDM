```python
import numpy as np
from pyspark import SparkConf, SparkContext
import time
```

```python
app_name = "hw2_q2"
master_URL = "spark://192.168.35.128:7077"
```

```python
start = time.time()
conf = SparkConf().setAppName(app_name).setMaster(master_URL)
sc = SparkContext(conf = conf)
```

```python
# 將資料以,分隔
def deal_process(data):
    header = data.first()
    data = data.filter(lambda x: x!=header)
    data = data.map(lambda x: x.split(","))
    return data
```

```python
# 計算平均並以list傳回
def cal(data, todo):
    sum_rdd = []
    sum_rdd.append(data[0])
    sum = float(data[144])
    if todo == "day":
        sum = sum / 2
    elif todo == "hour":
        sum = sum / 48
    sum_rdd.append(sum)
    return sum_rdd
```

```python
# 寫檔
def write(file_name, rdd_data, todo):
    output = open(file_name, "w")
    if todo == "day":
        header = '"IDLink", Average by day\n'
    elif todo == "hour":
        header = '"IDLink", Average by hour\n'
    output.write(header)
    output_format = "{},{}\n"
    for row in rdd_data.collect():
        output.write(output_format.format(row[0], row[1]))
```

```python
#讀入有關facebook的檔案並將其結合，計算每小時與每日平均
data_facebook_economy = deal_process(sc.textFile("Facebook_Economy.csv"))
data_facebook_microsoft = deal_process(sc.textFile("Facebook_Microsoft.csv"))
data_facebook_obama = deal_process(sc.textFile("Facebook_Obama.csv"))
data_facebook_palestine = deal_process(sc.textFile("Facebook_Palestine.csv"))
data_facebook = data_facebook_economy.union(data_facebook_microsoft).union(data_facebook_obama).union(data_facebook_palestine)
facebook_average_per_hour = data_facebook.map(lambda x: cal(x, "hour"))
facebook_average_per_day = data_facebook.map(lambda x: cal(x, "day"))
write("facebook_average_per_hour.csv", facebook_average_per_hour, "hour")
write("facebook_average_per_day.csv", facebook_average_per_day, "day")
```
